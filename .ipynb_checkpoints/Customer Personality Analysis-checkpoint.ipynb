```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
```

```python
df = pd.read_csv("marketing_campaign.csv", sep="\t")
df
```

```python
pd.set_option('display.max_columns', None)
df.describe()
```

```python
df.info()
```

```python
#As colunas "Z_CostContact" e "Z_Revenue" tem todos os valores iguais, a coluna "ID" não será relevante para as análises
df.drop(["Z_CostContact", "Z_Revenue", "ID"], axis=1, inplace=True)
df
```

```python
# Verificando se há linhas duplicadas
df[df.loc[:, "Year_Birth":"Income"].duplicated(keep=False)]

# Drop linhas duplicadas incluindo coluna "Response"
#df.drop_duplicates(inplace=True)

# Drop linhas duplicadas excluindo coluna "Response"
df.drop_duplicates(subset=["Year_Birth", "Education", "Marital_Status", "Income"], inplace=True)
df
```

```python
import matplotlib.pyplot as plt

df["Income"].plot();
```

```python
df[df["Income"] > 200000]
```

```python
# Verificamos um outliers na coluna Income e optamos por dropá-lo
df.drop(2233, inplace=True)
```

```python
df["Income"].plot();
```

```python
df["Year_Birth"].plot();
```

```python
df[df["Year_Birth"] < 1940]
```

```python
# Verificamos três outliers na coluna Year_Birth e optamos por dropá-los
df.drop([192, 239, 339], inplace=True)    # df = df[df["Year_Birth"] > 1940]
```

```python
df.groupby(["Education", "Marital_Status"])[["Income"]].describe()
```

```python
# Drop Alone, Absurd, YOLO
filtro = df[(df["Marital_Status"] == "Alone") | (df["Marital_Status"] == "Absurd") | (df["Marital_Status"] == "YOLO")]
df.drop(filtro.index, inplace=True)
```

```python
df.groupby(["Education", "Marital_Status"])[["Income"]].describe()
```

```python
# Verificando NaN
df.isna().sum()
df[df["Income"].isna()]
```

```python
df["Imputed_income"] = df["Income"].isna().astype('int64')
```

```python
df["Imputed_income"].sum()
```

```python
# Preencher NaN da coluna Income com média por Education
df["IncomeMean"] = df.groupby("Education")["Income"].transform('mean')
df
```

```python
df["Income"].fillna(df["IncomeMean"], inplace=True)
```

```python
df.isna().sum()
```

```python
df.groupby(["Education", "Marital_Status"])[["Income"]].describe()
```

```python
df_accepted = df.loc[:, "AcceptedCmp3":"Response"].drop("Complain", axis=1)
```

```python
df_accepted
```

```python
df_accepted["Education"] = df["Education"]
```

```python
df_accepted
```

```python
df1 = pd.DataFrame(df_accepted["Education"].value_counts()).reset_index()
df1.rename({"index": "Education", "Education": "EducationCount"}, axis=1, inplace=True)
df1
```

```python
df_accepted_e = df_accepted.groupby("Education").sum().reset_index().sort_values(["Education"])
df_accepted_e
```

```python
df_accepted_e = df_accepted_e.merge(df1, how='inner')
df_accepted_e
```

```python
(df_accepted_e[["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Response"]].div(df_accepted_e["EducationCount"], axis=0)*100).style.format('{0:,.2f}%')
```
