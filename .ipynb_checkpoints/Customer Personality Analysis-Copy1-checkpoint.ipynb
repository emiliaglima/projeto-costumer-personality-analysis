# Projeto - Análise da Personalidade do Cliente

### Introdução

A Análise da Personalidade do Cliente é uma análise detalhada dos clientes ideais de uma empresa. Ajuda uma empresa a entender melhor seus clientes e torna mais fácil para modificar seus produtos de acordo com suas necessidades, comportamentos e preocupações específicos de diferentes tipos de clientes. Neste trabalho extraímos informações de dados sobre 2240 clientes com 29 variáveis.

#### 1 - Analise do gasto total dos clientes em relação ao perfil dos grupos

#### 5 - Relação entre aceitação de campanha e escolaridade

```python
df_accepted = df.loc[:, "AcceptedCmp3":"Response"].drop("Complain", axis=1)
```

```python
df_accepted["Education"] = df["Education"]
```

```python
df1 = pd.DataFrame(df_accepted["Education"].value_counts()).reset_index()
df1.rename({"index": "Education", "Education": "EducationCount"}, axis=1, inplace=True)
df1
```

```python
df_accepted_e = df_accepted.groupby("Education").sum().reset_index().sort_values(["Education"])
```

```python
df_accepted_e = df_accepted_e.merge(df1, how='inner')
```

```python
df_campanha_edu = (df_accepted_e[["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Response"]].div(df_accepted_e["EducationCount"], axis=0)*100).style.format('{0:,.2f}%')
```

```python
df_campanha_edu
```

```python
df_campanha = df_campanha_edu.data
```

```python
plt.figure(figsize=(8,6))
x1 =  np.arange(1, 10, 2)
x2 = [x + 0.25 for x in x1]
x3 = [x + 0.25 for x in x2]
x4 = [x + 0.25 for x in x3]
x5 = [x + 0.25 for x in x4]
x6 = [x + 0.25 for x in x5]
coluna = ["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Response"]
x = [x1, x2, x3, x4, x5, x6]
color = ["darkorange", "lightblue", "palevioletred", "seagreen", "lightsteelblue", "moccasin"]

y1 = df_campanha.AcceptedCmp1.tolist()
plt.bar(x1, y1, width=0.25, label="C1", color="darkorange")

y2 = df_campanha.AcceptedCmp2.tolist()
plt.bar(x2, y2, width=0.25, label="C2", color="lightblue")

y3 = df_campanha.AcceptedCmp3.tolist()
plt.bar(x3, y3, width=0.25, label="C3", color="palevioletred")

y4 = df_campanha.AcceptedCmp4.tolist()
plt.bar(x4, y4, width=0.25, label="C4", color="seagreen")

y5 = df_campanha.AcceptedCmp5.tolist()
plt.bar(x5, y5, width=0.25, label="C5", color="lightsteelblue")

y6 = df_campanha.Response.tolist()
plt.bar(x6, y6, width=0.25, label="Last", color="moccasin")

# coloca o nome dos meses como label do eixo x
education = ['2n Cycle','Basic','Graduation','Master','PhD']
plt.xticks([x + 0.25 for x in range(1, 10, 2)], education)

# inseri uma legenda no gráfico
plt.legend()
plt.title("Aceitação nas campanhas publicitárias")
plt.show()
```

#### 6 - Relação entre aceitação de campanha e renda

```python
df_accepted_i = df_accepted.copy().drop("Education", axis=1)
```

```python
df_accepted_i["cut_income"] = df["cut_income"]
```

```python
df_accepted_i = df_accepted_i.groupby(["cut_income"]).sum().reset_index()
df_accepted_i
```

```python
df2 = pd.DataFrame(df["cut_income"].value_counts())
df2 = df2.reset_index()
df2.rename({"index": "cut_income", "cut_income": "IncomeCounts"}, axis=1, inplace=True)
df2
```

```python
df_accepted_i = df_accepted_i.merge(df2, how="inner", on=["cut_income"])
```

```python
df_accepted_i = (df_accepted_i[["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Response"]].div(df_accepted_i["IncomeCounts"], axis=0)*100).style.format('{0:,.2f}%')
```

```python
df_accepted_i = df_accepted_i.data
```

```python
df_accepted_i
```

```python
plt.figure(figsize=(8,6))
x1 =  np.arange(1, 8, 2)
x2 = [x + 0.25 for x in x1]
x3 = [x + 0.25 for x in x2]
x4 = [x + 0.25 for x in x3]
x5 = [x + 0.25 for x in x4]
x6 = [x + 0.25 for x in x5]
coluna = ["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Response"]
x = [x1, x2, x3, x4, x5, x6]
color = ["darkorange", "lightblue", "palevioletred", "seagreen", "lightsteelblue", "moccasin"]

y1 = df_accepted_i.AcceptedCmp1.tolist()
plt.bar(x1, y1, width=0.25, label="C1", color="darkorange")

y2 = df_accepted_i.AcceptedCmp2.tolist()
plt.bar(x2, y2, width=0.25, label="C2", color="lightblue")

y3 = df_accepted_i.AcceptedCmp3.tolist()
plt.bar(x3, y3, width=0.25, label="C3", color="palevioletred")

y4 = df_accepted_i.AcceptedCmp4.tolist()
plt.bar(x4, y4, width=0.25, label="C4", color="seagreen")

y5 = df_accepted_i.AcceptedCmp5.tolist()
plt.bar(x5, y5, width=0.25, label="C5", color="lightsteelblue")

y6 = df_accepted_i.Response.tolist()
plt.bar(x6, y6, width=0.25, label="Last", color="moccasin")

# coloca o nome dos meses como label do eixo x
income = ['(1500, 20000]','(20000, 50000]','(50000, 80000]','(80000, 165000]']
plt.xticks([x + 0.25 for x in range(1, 8, 2)], income)

# inseri uma legenda no gráfico
plt.legend()
plt.title("Aceitação nas campanhas publicitárias")
plt.show()
```
